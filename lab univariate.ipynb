```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset 
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')

# Display the first few rows of the dataset
print(df.head())

# Part 1: Understanding Product Categories
# 1 Generate a frequency table for the product category
category_frequency = df['category'].value_counts()

# Display the frequency table
print(category_frequency)

# Identify the top 5 most listed product categories
top_5_categories = category_frequency.head(5)

# Display the top 5 categories
print(top_5_categories)

```

```python
# 2 Visualizations 
# Bar chart for the distribution of products across different categories
plt.figure(figsize=(12, 6))
category_frequency.plot(kind='bar')
plt.title('Distribution of Products Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Number of Listings')
plt.show()

# Pie chart for the top 5 categories
plt.figure(figsize=(8, 8))
top_5_categories.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Top 5 Product Categories')
plt.ylabel('')
plt.show()
```

```python
# Part 2: Delving into Product Pricing
# Ensure the 'price' column is numeric (handling any non-numeric values or NaNs if necessary)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])

# 1 Measures of Centrality
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print("Mean price:", mean_price)
print("Median price:", median_price)
print("Mode price:", mode_price)

# 2 Measures of Dispersion
variance_price = df['price'].var()
std_deviation_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print("Variance of price:", variance_price)
print("Standard deviation of price:", std_deviation_price)
print("Range of price:", range_price)
print("Interquartile range (IQR) of price:", iqr_price)

# 3 Visualizations

# Histogram of product prices
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Box plot of product prices
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['price'])
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Part 3: Unpacking Product Ratings
# 1 Measures of Dispersion
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
interquartile_range = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print("Variance of stars:", variance_rating)
print("Standard deviation of stars:", std_dev_rating)
print("Interquartile range of stars:", interquartile_range)

# 2 Shape of the Distribution
skewness_stars = df['stars'].skew()
kurtosis_stars = df['stars'].kurt()

print("Skewness of stars:", skewness_stars)
print("Kurtosis of stars:", kurtosis_stars)

# 3 Visualizations
plt.figure(figsize=(10, 6))
sns.histplot(df['stars'], bins=20, kde=True)
plt.title('Distribution of Product ratings')
plt.xlabel('stars')
plt.ylabel('Frequency')
plt.show()
```
